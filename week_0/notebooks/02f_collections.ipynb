### Exercise F2 — Frequency map (medium)

````python
# Your turn
````

<details>
<summary><b>Solution</b></summary>

```python

```

### Exercise F3 — Nested structures (hard)
Create a class roster as list-of-dicts and compute average age.

````python
# Your turn
````

<details>
<summary><b>Solution</b></summary>

```python
students=[{"name":"Ali","age":22},{"name":"Sara","age":24}]

print(round(avg,2))
```
